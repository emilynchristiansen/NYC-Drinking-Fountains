```python
import geopandas as gpd
```

```python
osm_data = gpd.read_file("prospect_park_amenities_drinking_water.geojson")
city_data = gpd.read_file("NYC_Parks_Drinking_Fountains_Prospect_Park (1).geojson")
```

```python
print("OSM CRS:", osm_data.crs)
print("City CRS:", city_data.crs)
print("OSM count:", len(osm_data))
print("City count:", len(city_data))
```

```python
osm_proj = osm_data.to_crs(epsg=3857)
city_proj = city_data.to_crs(epsg=3857)
```

```python
matched = gpd.sjoin_nearest(osm_proj, city_proj, how="inner", max_distance=20, distance_col="distance_m")

print("Matched count:", len(matched))
```

```python
# OSM points with no city match
osm_only = osm_proj[~osm_proj.index.isin(matched.index)]

# City points with no OSM match
city_only = city_proj[~city_proj.index.isin(matched["index_right"])]

print("OSM only:", len(osm_only))
print("City only:", len(city_only))
```
